**PC Games Sales Analysis using python.**

```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv('vgsales(original).csv')
df.info()
```

```python
df.head(100)
```

```python
df.dropna()

```

```python
df["Year"].fillna((df["Year"].median()), inplace=True)
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("darkgrid")
```

```python
df.describe().transpose()
```

```python
plt.figure(figsize=(12,6), dpi=150)
sns.scatterplot(data=df,
                x="Year",
                y="Global_Sales")

plt.show()
```

```python
plt.figure(figsize=(12,6), dpi=150)
sns.countplot(data=df,
              x="Genre",
              order = df["Genre"].value_counts().index,
              palette="Dark2")
plt.xticks(rotation=90)
plt.show()
```

Top 10 games making most sales globally.

```python
df.nlargest(10, "Global_Sales")
```

Top 5 gaming genres that are making high sales.

```python
sales_genre = df.groupby("Genre").agg({"Global_Sales": pd.Series.sum})
sales_genre.nlargest(5, "Global_Sales")
```

Which game made the most sales per region and globally?

```python
print(f"NA: {df.nlargest(1, 'NA_Sales').values[0,1]}")
print(f"EU: {df.nlargest(1, 'EU_Sales').values[0,1]}")
print(f"JP: {df.nlargest(1, 'JP_Sales').values[0,1]}")
print(f"Other: {df.nlargest(1, 'Other_Sales').values[0,1]}")
print(f"Global: {df.nlargest(1, 'Global_Sales').values[0,1]}")
```

Which publishers made the most sales?

```python
sales_publisher = df.groupby("Publisher").agg({"Global_Sales": pd.Series.sum})
sales_publisher.nlargest(5, "Global_Sales")
```

Boxplot

```python
sns.boxplot(x='Genre',y='Year', data=df)
plt.figure(figsize=(30,60))
```
